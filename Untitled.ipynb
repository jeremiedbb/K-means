```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ch_d = pd.read_csv('src/chunked_loop.csv')
ch_ = pd.read_csv('src/chunked_loop_a.csv')
ch_1 = pd.read_csv('src/chunked_1.csv')
ch_05 = pd.read_csv('src/chunked_05.csv')
nch = pd.read_csv('src/loop.csv')
x = np.array(ch_d['size'])
```

```python
ax = plt.subplot(111)
ax.plot(x, np.array(ch_d['time']), label='data + chunked')
ax.plot(x, np.array(ch_['time']), label='chunked')
ax.plot(x, np.array(ch_1['time']), label='chunked 1M')
ax.plot(x, np.array(ch_05['time']), label='chunked 0.5M')
ax.plot(x, np.array(nch['time']), label='not chunked')
plt.xlabel('size (M)')
plt.ylabel('time (s)')
plt.title('')
plt.legend()
plt.show()
```

```python
df = pd.read_csv('src/chunked_vs_size.csv')
ax = plt.subplot(111)
ax.plot(df['size'], df['time'])
plt.xlabel('chunk size (M)')
plt.ylabel('time (s)')
#plt.title('n_sample: 100000 | n_feature: 3 | n_component: 1024')
plt.show()
```
