```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wm = pd.read_csv('bench_kmeans_wm_32cores.csv')
wm.plot(x='wm', y='time').set_xscale('log')
plt.xlabel('working memory')
plt.ylabel('time')
plt.title('time vs working_memory')
plt.show()
```

```python
wm2 = pd.read_csv('bench_kmeans_wm_4cores_AVX.csv')
wm2.plot(x='wm', y='time').set_xscale('log')
plt.xlabel('working memory')
plt.ylabel('time')
plt.title('time vs working_memory | 4 cores | AVX')
plt.show()
```
