```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./src/bench_impl.py')

df1 = df[df['impl']=='intel'].reset_index(drop=True)
df1.rename(columns={'t':'intel'}, inplace=True)
df1.drop(['impl'], axis=1, inplace=True)

df2 = df[df['impl']=='PR'].reset_index(drop=True)
df2.rename(columns={'t':'PR'}, inplace=True)
df2.drop(['impl'], axis=1, inplace=True)

df3 = df[df['impl']=='master'].reset_index(drop=True)
df3.rename(columns={'t':'master'}, inplace=True)
df3.drop(['impl'], axis=1, inplace=True)

df4 = df[df['impl']=='PR_t'].reset_index(drop=True)
df4.rename(columns={'t':'PR_t'}, inplace=True)
df4.drop(['impl'], axis=1, inplace=True)

df5 = df[df['impl']=='PR_tt'].reset_index(drop=True)
df5.rename(columns={'t':'PR_tt'}, inplace=True)
df5.drop(['impl'], axis=1, inplace=True)

df = df1.merge(df2, on=['ns','nc','nf'])
df['x.'] = df['PR'] / df['intel']
df['x.'] = df['x.'].apply(lambda x: np.around(x, 1))

df = df.merge(df3, on=['ns','nc','nf'])
df['x..'] = df['master'] / df['intel']
df['x..'] = df['x..'].apply(lambda x: np.around(x, 1))

df
```
