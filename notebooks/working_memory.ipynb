```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../bench_kmeans_wm.csv')
means = df.groupby('wm').mean().reset_index()
stds = df.groupby('wm').std().reset_index()
fig = plt.figure()
ax = fig.add_subplot(111)
ax.errorbar(means['wm'], means['time'], yerr=stds['time'], )
ax.set_xscale('log')
ax.set_title('KMeans performance vs working memory (24Mo cpu cache)')
ax.set_xlabel('working memory (Mo)')
ax.set_ylabel('time (s)')
plt.show()
```

```python
df = pd.read_csv('../bench_kmeans_wm_4.csv')
df = df[df['wm']!=0.1]
means = df.groupby('wm').mean().reset_index()
stds = df.groupby('wm').std().reset_index()
fig = plt.figure()
ax = fig.add_subplot(111)
ax.errorbar(means['wm'], means['time'], yerr=stds['time'], )
ax.set_xscale('log')
ax.set_title('KMeans performance vs working memory (4Mo cpu cache)')
ax.set_xlabel('working memory (Mo)')
ax.set_ylabel('time (s)')
plt.show()
```

```python
df = pd.read_csv('../bench_working_memory.csv')
dfp = df[df['n_samples']==40000].groupby('working_memory').mean().reset_index()
dfp = dfp[dfp['working_memory'] != 1]
dfp = dfp[dfp['working_memory'] != 2]
dfp = dfp[dfp['working_memory'] != 4]
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(dfp['working_memory'], dfp['elapsed'])
ax.set_xscale('log')
ax.set_title('dtype = np.float64')
ax.set_xlabel('working memory (Mo)')
ax.set_ylabel('time (s)')
plt.show()
#dfp.plot(x='working_memory', y='elapsed').set_xscale('log')
#plt.show()
```
