```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bench_chunked_loop.csv')
df['chunk_size'] *= 4/1000000
means = df.groupby('chunk_size').mean().reset_index()
stds = df.groupby('chunk_size').std().reset_index()
fig = plt.figure()
ax = fig.add_subplot(111)
ax.errorbar(means['chunk_size'], means['time1'], yerr=stds['time1'], label='memoryview')
ax.errorbar(means['chunk_size'], means['time2'], yerr=stds['time2'], label='buffer')
ax.set_xscale('log')
ax.set_title('Loop over 20Mo 1D array')
ax.set_xlabel('chunk_size (Mo)')
ax.set_ylabel('time (s)')
ax.legend()
plt.show()
```
